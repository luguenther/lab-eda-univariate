```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

# Load the dataset
file_path = "/Users/lukasgunther/Desktop/Ironhack/Bootcamp/Week5/Day1/Labs/lab-eda-univariate/amz_uk_price_prediction_dataset.csv"
df = pd.read_csv(file_path)

# Display basic dataset info
print("Dataset Info:")
df.info()
print("\nFirst few rows of the dataset:")
print(df.head())
```

```python
### PART 1: UNDERSTANDING PRODUCT CATEGORIES ###

# 1️⃣ Frequency Table for Product Categories
category_counts = df['category'].value_counts()
top_5_categories = category_counts.head(5)

print("\nTop 5 Most Listed Product Categories:")
print(top_5_categories)
```

```python
# 2️⃣ Bar Chart - Distribution of Product Categories
plt.figure(figsize=(12,6))
sns.barplot(x=top_5_categories.index, y=top_5_categories.values, palette="viridis")
plt.xticks(rotation=45)
plt.xlabel("Category")
plt.ylabel("Number of Listings")
plt.title("Top 5 Most Listed Product Categories")
plt.show()
```

```python
# 3️⃣ Pie Chart - Proportions of Top Categories
plt.figure(figsize=(8,8))
top_5_categories.plot.pie(autopct='%1.1f%%', colors=sns.color_palette("viridis", 5))
plt.title("Proportion of Listings by Category")
plt.ylabel("")
plt.show()
```

```python
### PART 2: DELVING INTO PRODUCT PRICING ###

# 1️⃣ Measures of Centrality
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]

print(f"\nMean Price: {mean_price}")
print(f"Median Price: {median_price}")
print(f"Mode Price: {mode_price}")
```

```python
# 2️⃣ Measures of Dispersion
variance_price = np.var(df['price'], ddof=1)
std_dev_price = np.std(df['price'], ddof=1)
range_price = df['price'].max() - df['price'].min()
iqr_price = np.percentile(df['price'], 75) - np.percentile(df['price'], 25)

print(f"\nPrice Variance: {variance_price}")
print(f"Price Standard Deviation: {std_dev_price}")
print(f"Price Range: {range_price}")
print(f"Interquartile Range (IQR): {iqr_price}")
```

```python
# 3️⃣ Histogram - Price Distribution
plt.figure(figsize=(10,6))
sns.histplot(df['price'], bins=50, kde=True, color='blue')
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Distribution of Product Prices")
plt.show()
```

```python
# 4️⃣ Boxplot - Identifying Outliers
plt.figure(figsize=(8,5))
sns.boxplot(x=df['price'], color='blue')
plt.xlabel("Price")
plt.title("Boxplot of Product Prices")
plt.show()
```

```python
### PART 3: UNPACKING PRODUCT RATINGS ###

# 1️⃣ Measures of Centrality
mean_rating = df['stars'].mean()
median_rating = df['stars'].median()
mode_rating = df['stars'].mode()[0]

print(f"\nMean Stars: {mean_rating}")
print(f"Median Stars: {median_rating}")
print(f"Mode Stars: {mode_rating}")
```

```python
# 2️⃣ Measures of Dispersion
variance_rating = np.var(df['stars'], ddof=1)
std_dev_rating = np.std(df['stars'], ddof=1)
iqr_rating = np.percentile(df['stars'], 75) - np.percentile(df['stars'], 25)

print(f"\nRating Variance: {variance_rating}")
print(f"Rating Standard Deviation: {std_dev_rating}")
print(f"Rating Interquartile Range (IQR): {iqr_rating}")
```

```python
# 3️⃣ Skewness & Kurtosis (Shape of Distribution)
skewness = skew(df['stars'].dropna())
kurt = kurtosis(df['stars'].dropna())

print(f"\nSkewness of Ratings: {skewness}")
print(f"Kurtosis of Ratings: {kurt}")
```

```python
# 4️⃣ Histogram - Rating Distribution
plt.figure(figsize=(10,6))
sns.histplot(df['stars'].dropna(), bins=20, kde=True, color='green')
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.title("Distribution of Product Ratings")
plt.show()
```
